```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#1-Créer Un DataFrame en utilisant les données de fichier ‘’market_basket.txt’qui contient les identifiants des caddies et les produits associées.
data=pd.read_csv('market_basket.txt',sep='\t')
# 2-Afficher les 10 premières lignes du DataFrame.
data.head(10)
```

```python
# 4-Écrire un script python qui permet de Construire un table binaire indiquant la présence de chaque produit au niveau des caddies (True:1 si le produit est présent dans le caddie et 0 dans le cas réciproque).
from os import sep
col=data['Product'].sort_values().unique()
lig=data['ID'].unique()

matrix=np.zeros((len(lig),len(col)))
k=0
for i in lig:
  df=data[data["ID"]==i]
  a=0
  for j in col:
    if df['Product'].str.contains(j).any():
      matrix[k][a]=1
    a=a+1
  k=k+1

df = pd.DataFrame(matrix, index=lig,columns=col)
df
```

```python
# 7-Écrire un script python de la fonction a_priori() qui permet l’extraction des itemsets les plus fréquents. ( on définit un min_supp=0.025 et un longueur maximum de 4 produits)
import itertools
nbr_max_produit=2
def ExtractItemset(df, n):
    return list(itertools.combinations(df, n))

def CalculSupport (df,subsets,n):
  tab=[]
  k=0
  for i in subsets:
    subset=df[list(i)]
    w=subset[subset.sum(axis='columns')==n].count()
    tab.append(w[0])
    k=k+1
  return(tab)
#main
min_supp=0.025*len(lig)
for i in range(1,nbr_max_produit+1):
  itemset=ExtractItemset(col,i)
  Support=CalculSupport(df,itemset,i)
  ff= pd.DataFrame(itemset)
  sup=pd.DataFrame(Support)
  sup.columns=['Support']
  frame=[ff,sup]
  F=pd.concat(frame,axis=1)
  C=F[F["Support"]>min_supp]
  print("C",i)
  print(C)
```

```python
def Regle(item):
  it=item[:-1]
  for i in range(1,nbr_max_produit):
    List=ExtractItemset(it,i)
    PremisList=List[:-i]
    CclList=List[i:]
    PremisSupport=CalculSupport(df,PremisList,i)
    TotalSupport=CalculSupport(df,[it],nbr_max_produit)
  return(PremisList,CclList,np.array(TotalSupport)/np.array(PremisSupport))



AprioriRes=[]
k=0
for i in C.index:
  k=k+1
  ligne=list(C.loc[i,:])
  Premis,Ccl,Confidence=Regle(ligne)
  print('Regle :',Premis,' -> ',Ccl,' With confidence',Confidence)
```

```python
#8-Afficher les 15 premiers itemsets.
freq_itemsets = apriori(df,min_support=0.025,max_len=4,use_colnames=True)
print(freq_itemsets.head(15))
```

```python
# 9-Ecrire une fonction is_inclus() qui permet de vérifier si un sous-ensemble items est inclus dans l’ensemble x.
def is_inclus(_subset,_set):
  if len(_subset)>len(_set):
    return(False)
  for i in _subset:
    if not(i in _set):
      return(False)
  return(True)

```

```python
# 10- Afficher les itemsets comprenant le produit ‘Aspirin’.
import numpy
def is_inclus(x,items):
  return items.issubset(x)
id = numpy.where(freq_itemsets.itemsets.apply(is_inclus,items={'Aspirin'}))
print(freq_itemsets.loc[id])
```

```python
# 11-Afficher les itemsets contenant Aspirin et Eggs.
print(freq_itemsets[freq_itemsets['itemsets'].ge({'Aspirin','Eggs'})])
```

```python
#12-Nous produisons les règles à partir des itemsets fréquents. Elles peuvent être très nombreuses,nous en limitons la prolifération en définissant un seuil minimal (min_threshold = 0.75) sur une mesure d’intérêt, en l’occurrence la confiance dans notre exemple (metric = ‘’confidence’’). 
regles = association_rules(freq_itemsets,metric="confidence",min_threshold=0.75)
```
